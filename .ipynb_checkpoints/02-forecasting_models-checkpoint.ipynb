In this dataset we take a look at several datasets containing statistics about wine market in the United States.

## Load and Inspect Data
For this project we analyse data from multiple sources. We look at the statistics related to the wine industry in the US. The datasets considered include:
* **Gross margin of premium wineries in the United States 2002-2022:** This statistic shows the gross margin of premium wineries in the United States from 2002 to 2022
* **Market share of the U.S. alcohol industry by beverage 2000-2022:** Sales market share of the United States alcohol industry from 2000 to 2022, by beverage.
* **Total wine consumption of the U.S. 1934-2022**
* **U.S. and Californian wine production 2006-2022:** Wine production in the United States and in California from 2006 to 2022 (in million gallons.
* **Sales volume of wine in the U.S. 1999-2021:** Sales volume of wine in the United States from 1999 to 2021 ( in million 9-liter cases).
* **Californian wine market value in the U.S. 2006-2022:** Retail value of the Californian wine market in the United States from 2006 to 2022 (in billion U.S. dollars)

#### Load Data

```python
gross_margin_premium = pd.read_excel(
    "dataset/market_stats/statistic_id209163_gross-margin-of-premium-wineries-in-the-united-states-2002-2022.xlsx", 
    sheet_name='Data', skiprows=4,
    names=["0:Unnamed", "Year", "Gross Margins (%)", "1:Unnamed"],
    usecols=["Year", "Gross Margins (%)"]
)
gross_margin_premium.shape
```

```python
beverage_share = pd.read_excel(
    "dataset/market_stats/statistic_id233699_market-share-of-the-us-alcohol-industry-by-beverage-2000-2022.xlsx", 
    sheet_name='Data', skiprows=4,
    names=["0:Unnamed", "Year", "Spirits", "Wine", "Beer", "1:Unnamed"],
    usecols=["Year", "Spirits", "Wine", "Beer"]
).rename(columns={"Spirits": "Spirits Market Shares (%)", "Wine": "Wine Market Shares (%)", "Beer": "Beer Market Shares (%)"})
beverage_share.shape
```

```python
wine_consumption = pd.read_csv("dataset/market_stats/wine_consumption.tsv", delimiter="\t").rename(columns={
    "Total Wine per Resident (gals)": "Consumption per Resident (Mgal)",
    "Total Wine (Mgal)": "Wine Consumption (Mgal)",
    "Total Table Wine (Mgal)": "Table Wine Consumption (Mgal)"
})
wine_consumption.shape
```

```python
wine_production = pd.read_excel(
    "dataset/market_stats/statistic_id259493_us-and-californian-wine-production-2006-2022.xlsx", 
    sheet_name='Data', skiprows=4,
    names=["0:Unnamed", "Year", "US (in million gallons)", "California (in million gallons)"],
    usecols=["Year", "US (in million gallons)", "California (in million gallons)"]
).rename(columns={"US (in million gallons)": "US Wine Production (Mgal)", "California (in million gallons)": "California Wine Production (Mgal)"})
wine_production.shape
```

```python
sales_volumes = pd.read_excel(
    "dataset/market_stats/statistic_id715474_sales-volume-of-wine-in-the-us-1999-2021.xlsx", 
    sheet_name='Data', skiprows=4,
    names=["0:Unnamed", "Year", "Volume (in million 9-liter cases)"],
    usecols=["Year", "Volume (in million 9-liter cases)"]
).rename(columns={"Volume (in million 9-liter cases)": "US Sales Volume (9-liter cases)"})
sales_volumes.shape
```

```python
californian_market_value = pd.read_excel(
    "dataset/market_stats/statistic_id737296_californian-wine-market-value-in-the-us-2006-2022.xlsx", 
    sheet_name='Data', skiprows=4,
    names=["0:Unnamed", "Year", "Value (in billion U.S. dollars)"],
    usecols=["Year", "Value (in billion U.S. dollars)"]
).rename(columns={"Value (in billion U.S. dollars)": "Californian Wine Market Value ($Bn)"})
californian_market_value.shape
```

Correct inconsistencies in `Year` column

```python
gross_margin_premium["Year"] = gross_margin_premium["Year"].str.replace("*", "", regex=False).astype(int)
```

A decade is missing in the beverage market share data. We go from 2000 straight to 2010. We may discard the 2000 row.

```python
beverage_share = beverage_share[beverage_share["Year"] >= 2010]
```

#### Merge Datasets

```python
df_merged = pd.merge(gross_margin_premium, beverage_share, on="Year", how="outer")
df_merged = pd.merge(df_merged, wine_consumption, on="Year", how="outer")
df_merged = pd.merge(df_merged, wine_production, on="Year", how="outer")
df_merged = pd.merge(df_merged, sales_volumes, on="Year", how="outer")
df_merged = pd.merge(df_merged, californian_market_value, on="Year", how="outer")
df_merged = df_merged.sort_values(by="Year")
df_merged.shape
```

Ensure variables are using the same metrics when possible

```python
def convert_cases_to_gallons(million_cases):
    gallons_per_case = 2.37754858  # Conversion factor from 9-liter case to gallons
    million_gallons = million_cases * gallons_per_case
    return million_gallons
```

```python
df_merged["US Sales Volume (Mgal)"] = df_merged["US Sales Volume (9-liter cases)"].apply(convert_cases_to_gallons)
```

## Exploratory Data Analysis

```python
df_merged.describe()
```

### Trend Analysis
* **Gross Margins:** Average of 55.08%, with a standard deviation of 2.54%, indicating relatively stable margins over the years.
* **Market Shares:** Wine has an average market share of 16.78%, while spirits and beer have 36.16% and 47.03%, respectively. This indicates a competitive alcoholic beverage market.
* **Consumption and Production:** Wine consumption per resident averages 1.65 million gallons, while total wine consumption averages 421 million gallons. US and California wine production are significantly lower in available data points but are crucial for understanding supply trends.
* **Sales Volume and Market Value:** `US Sales Volume (9-liter cases)` and `Californian Wine Market Value` have fewer data points, but they are important for financial forecasting and market analysis.

```python
fig,axes = plt.subplots(figsize=(15.5,5.5), ncols=4, nrows=2)
sns.lineplot(data=df_merged, x="Year", y="Gross Margins (%)", ax=axes[0][0]).set_title("Gross Margins (%)")
sns.lineplot(data=df_merged, x="Year", y="Consumption per Resident (Mgal)", ax=axes[0][1]).set_title("Consumption per Resident (Mgal)")
sns.lineplot(data=df_merged, x="Year", y="Wine Consumption (Mgal)", ax=axes[0][2], label="Total Wine Consumption")
sns.lineplot(data=df_merged, x="Year", y="Table Wine Consumption (Mgal)", ax=axes[0][2], label="Table Wine Consumption")
axes[0][2].set_title("Wine Consumption (Mgal)")
sns.lineplot(data=df_merged, x="Year", y="US Wine Production (Mgal)", ax=axes[0][3]).set_title("US Wine Production (Mgal)")
sns.lineplot(data=df_merged, x="Year", y="US Sales Volume (9-liter cases)", ax=axes[1][0]).set_title("US Sales Volume (9-liter cases)")
sns.lineplot(data=df_merged, x="Year", y="Californian Wine Market Value ($Bn)", ax=axes[1][1]).set_title("Californian Wine Market Value ($Bn)")
sns.lineplot(data=df_merged, x="Year", y="California Wine Production (Mgal)", ax=axes[1][2]).set_title("California Wine Production (Mgal)")
# Beverage market share
axes[1][3].bar(beverage_share["Year"], beverage_share["Wine Market Shares (%)"], label='Wine')
axes[1][3].bar(beverage_share["Year"], beverage_share["Spirits Market Shares (%)"], bottom=beverage_share["Wine Market Shares (%)"], label='Spirits')
axes[1][3].bar(beverage_share["Year"], beverage_share["Beer Market Shares (%)"], bottom=beverage_share["Wine Market Shares (%)"] + beverage_share["Spirits Market Shares (%)"], label='Beer')
axes[1][3].set_title("Market Share (%)")
axes[1][3].legend()
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=4))
plt.tight_layout()
```

### Correlation Analysis
* `US Sales Volume (9-liter cases)` and `Wine Consumption (Mgal)`: Extremely high correlation (0.996) indicates that sales volume is closely tied to wine consumption.
* `Consumption per Resident (Mgal)` and `US Sales Volume (Mgal)`: Also very high correlation (0.995) suggests that per capita consumption is a strong indicator of overall sales.
* `US Wine Production` and `California Wine Production`: High correlation (0.940) shows that California's production trends are closely aligned with overall US production.
* `US Sales Volume` and `Beer Market Shares (%)`: Strong negative correlation (-0.980) indicates that as wine sales volume increases, beer market share tends to decrease. This suggests a competitive dynamic between beer and wine markets.
* `Gross Margins` and `Consumption per Resident`: Positive correlation (0.793) indicates that higher per capita wine consumption is associated with better gross margins.

```python
correlation_matrix = df_merged.corr(method='pearson')
```

```python
plt.figure(figsize=(9,7))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f")
plt.title('Correlation Matrix')
#plt.savefig("figs/correlation_matrix.png", dpi=300, facecolor='w', bbox_inches = 'tight')
plt.show()
```
