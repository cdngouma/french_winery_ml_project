```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

import re

from sklearn.manifold import TSNE
```

In this notebook, we have a look and analyze the dataset `marketing_campaign` found on [Kaggle](https://www.kaggle.com/datasets/rodsaldanha/arketing-campaign) which contains information about a marketing campaign. The original goal of this dataset is to predict how a customer would respond to an offer for a given product or service. The dataset has 29 columns and 2240 entries in total. The data was published in 2019 on [Kaggle](https://www.kaggle.com/datasets/rodsaldanha/arketing-campaign).

The context of our project is to study the wine market in California and the broader USA. Given this problem, we consider demographic information.
* `ID`: Unique identification of a customer
* `Year_Birth`: Year of birth of the customer
* `Education`: Customerâ€™s level of education
* `Marital_Status`: Marital status
* `Income`: Annual Income
* `Kidhome`: Number of small children in the home
* `Teenhome`: Number of teenagers in the home
* `MntWines`: Amount spent on wine in the past 2 years

Findings:
* Out of the columns considered, `Customer_Age`, `Education`, `Income`, and `Kidhome` have a correlation with the amount spent on wine `MntWines`.
* A large number of customers (~56%) are minimal consumers, meaning that they spent at most $250 on wine over 2 years.
* Seniors customers (60+) tend to soend more on wine than younger customers.
* Customers with PhD also tend to spend more on wine than those with undergraduate or Master's level of education.
* Income positively correlates with the amount spent on wine. The higher the income, the more customers are able to spend on wine. However, the range remains quite wide.
* Having a small child in the home has a correlation on the amount spent on wine. Customers with no children spend a significantly larger amount on wine than those without small children.
* A visualization of the dimensionally reduced data shows 2-4 potential clusters of customers with different spending habits. We could thus use a clustering algorithm to automatically identify the clusters and study the similarities between customers in a given cluster.

## Helper functions

```python
def t_test(x1, x2):
    t_stat, p_value = stats.ttest_ind(x1, x2, equal_var=False)
    print(f"T-statistic: {np.round(t_stat, 10)}")
    print(f"P-value: {np.round(p_value, 10)}")
```

Here we split the customers based on their age in 3 groups: `Young Adults`, `Adults`, and `Seniors`. 

```python
def get_age_group(age):
    if age < 0:
        raise Error("Negative age is invalid")
    if age < 18:
        return "Underaged"
    elif age < 35:
        return "Young Adults"
    elif age < 65:
        return "Adults"
    else:
        return "Seniors"
    
def get_age_group2(age):
    if age < 0:
        raise Error("Negative age is invalid")
    elif age <= 60:
        return "Adults"
    else:
        return "Seniors"
    
def get_generation(age):
    if age < 0:
        raise Error("Negative age is invalid")
    if age >= 59:
        return "Baby Boomers & Older"
    elif age >= 41:
        return "Generation X"
    elif age >= 10:
        return "Millennials & Generation Z"
    else:
        raise Error("Too young")
```

```python
df_clean["Customer_Age_Group"] = df_clean["Customer_Age"].apply(lambda x: get_age_group(int(x)))
```

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df_clean, x="Customer_Age_Group")
```

A boxplot of the data shows an extreme outlier. While most values of `Income` are between 0 and 160000, there one entry with an income of 666666. This value also seems quite odd as an annual income for an individual and could be the result of an error. As it is the only instance, we can safely remove it from the dataset.

```python
def indentify_outliers(arr):
    q1 = np.percentile(arr, 25)
    q3 = np.percentile(arr, 75)
    
    iqr = q3 - q1
    threshold = 1.5

    outliers = arr[(arr < q1 - threshold * iqr) | (arr > q3 + threshold * iqr)]
    
    return outliers
```

```python
income_outliers = indentify_outliers(df_clean["Income"].values)
```

We remove outliers

```python
df_clean = df_clean[~df_clean["Income"].apply(lambda x: x in income_outliers)]
```

### `Amount_Group`

Let's perform a manual clustering of the data to study difference between potential subgroups.

```python
def categorize_amount(amount):
    if amount < 0:
        raise Error("Negative amount")
    
    if amount == 0:
        return np.nan
    elif amount <= 40:
        return "0-40"
    elif amount <= 100:
        return "41-100"
    elif amount <= 300:
        return "101-300"
    elif amount <= 1000:
        return "301-1000"
    else:
        return "1000+"
```

```python
df_clean["Amount_Group"] = df_clean["MntWines"].apply(categorize_amount)
```

```python
cat_order = ["0-40", "41-100", "101-300", "301-1000", "1000+"]
df_clean["Amount_Group"] = pd.Categorical(df_clean['Amount_Group'], categories=cat_order, ordered=True)
```

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df_clean.sort_values(by="Amount_Group"), x="Amount_Group")
```

## Data Preprocessing

```python
def preprocess_data(data):
    # Select relevant columns
    columns = ["ID", "Year_Birth", "Education", "Income", "Kidhome", "MntWines"]
    df_clean = data[columns].copy()
    
    # Remove non spenders
    df_clean = df_clean[df_clean["MntWines"] > 0]
    # Extract and filter out age dead people
    df_clean["Customer_Age"] = 2019 - df_clean["Year_Birth"]
    #df_clean = df_clean[df_clean["Customer_Age"] <= 80]
    # Add age groups
    df_clean["Customer_Generation"] = df_clean["Customer_Age"].apply(lambda x: get_generation(int(x)))
    df_clean["Customer_Age_Group"] = df_clean["Customer_Age"].apply(lambda x: get_age_group(int(x)))
    df_clean["Customer_Age_Group2"] = df_clean["Customer_Age"].apply(lambda x: get_age_group2(int(x)))
    # Ramove customers with Basic education
    df_clean = df_clean[df_clean["Education"] != "Basic"]
    df_clean["Education2"] = df_clean["Education"].replace({
        "Graduation": "Undergraduate & Master", "2n Cycle": "Undergraduate & Master", "Master": "Undergraduate & Master"
    })
    df_clean["Education"] = df_clean["Education"].replace({"Graduation": "Undergraduate", "2n Cycle": "Undergraduate"})
    # Filter out income outliers
    df_clean = df_clean[df_clean["Income"] < 666666]
    # Make Kidhome binary
    df_clean["Kidhome2"] = (df_clean["Kidhome"] > 0).astype(int)
    # Add amount group
    df_clean["Amount_Group"] = df_clean["MntWines"].apply(categorize_amount)
    cat_order = ["0-40", "41-100", "101-300", "301-1000", "> 1000"]
    df_clean["Amount_Group"] = pd.Categorical(df_clean['Amount_Group'], categories=cat_order, ordered=True)
    # Remove income outliers
    income_outliers = indentify_outliers(df_clean["Income"].values)
    df_clean = df_clean[~df_clean["Income"].apply(lambda x: x in income_outliers)]
    
    return df_clean
```

```python
preprocess_data(data).to_csv("dataset/marketing_campaign_clean.csv", index=False)
```
