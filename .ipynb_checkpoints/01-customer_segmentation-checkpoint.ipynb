In this notebook, we have a look and analyze the dataset `marketing_campaign` found on [Kaggle](https://www.kaggle.com/datasets/rodsaldanha/arketing-campaign) which contains information about a marketing campaign. The original goal of this dataset is to predict how a customer would respond to an offer for a given product or service. The dataset has 29 columns and 2240 entries in total. The data was published in 2020 on [Kaggle](https://www.kaggle.com/datasets/rodsaldanha/arketing-campaign).

The context of our study is analysing the US wine market, our primary focus should be on demographic, behavioral, and engagement data to identify distinct customer segments.
* `Year_Birth`, `Education`, `Marital_Status`, `Income`: Demographic information of the customers
* `Kidhome`, `Teenhome`: Household information
* `Dt_Customer`: The date when the customer was enrolled, which can be used to calculate customer tenure
* `Recency`: The number of days since the last purchase, indicating customer engagement and loyalty.
* `MntWines`: Amount spent on wine
* `MntFruits`, `MntMeatProducts`, `MntFishProducts`, `MntSweetProducts`, `MntGoldProds`: Amount spent on other products such as fruits, meat, fish, sweets, and gold.
* `NumDealsPurchases`: Number of purchases made with a discount
* `NumCatalogPurchases`: Number of purchases using a catalog
* `NumStorePurchases`: Number of purchases made in-store
* `NumWebVisitsMonth`: Number of visits to the company's website
* `Complain`: Indicates if the customer has lodged a complaint

## 3. Data Cleaning and Preparation

#### Numerical Variables

Given the dataset was published in 2017, we create a new column `Age` by subtracting the `Year_Birth` from 2017. We then remove dead people from the dataset. We do not need to remove underage people as the drinking age is 21 in the US (which is the lowest age in the dataset).

```python
df_relevant["Age"] = 2020 - df_relevant["Year_Birth"]
```

```python
df_relevant["Age"].describe()
```

```python
df_relevant = df_relevant[df_relevant["Age"] <= 85].copy()
```

Some income do not make much sense such `666666`. Aside from this outlier, all the incomes fall between \\$1730 and \\$156924. Around 1% of customers have an income of $10,000/year or less. This is could indicate part-time workers, retirees, or individuals with limited sources of income. Although making up a small percentage of the dataset, removing these custimers could lead to a biased analysis.

```python
df_relevant["Income"].describe()
```

```python
sns.boxplot(data=df_relevant, x="Income")
```

```python
df_relevant = df_relevant[df_relevant["Income"] != 666666]
```

We convert `Dt_Customer` to appropriate data type then create `Customer_Tenure` from the `Dt_Customer` column. This refers to the length of time a customer has been associated with the marketing company. As we don't have the exact date but just the year, we assume the dataset was published in January 2020.

```python
df_relevant["Dt_Customer"] = pd.to_datetime(df_relevant["Dt_Customer"], dayfirst=True)
```

```python
df_relevant["Customer_Tenure (days)"] = (pd.to_datetime("01-01-2020") - df_relevant["Dt_Customer"]).dt.days
```

We also compute the total number of purchases made by the customers on all plateforms

```python
df_relevant["TotalNumPurchases"] = df_relevant["NumCatalogPurchases"] + df_relevant["NumWebPurchases"] + df_relevant["NumStorePurchases"]
```

#### Categorical Variables

`Marital_Status` has a few inconsistencies. We find labels such as "YOLO", "Alone", and "Absurd". We can replace them by "Single", which is more appropriate.

```python
df_relevant["Marital_Status"].value_counts()
```

```python
df_relevant["Marital_Status"] = df_relevant["Marital_Status"].replace({"Alone": "Single", "Absurd": "Single", "YOLO": "Single"})
```

Additionally, we can combine the household data to create a `HouseholdSize` column. We make sure to add 2 if the customer is married and 1 if they are single.

```python
df_relevant["Household_Size"] = df_relevant["Kidhome"] + df_relevant["Teenhome"]
df_relevant["Household_Size"] += df_relevant["Marital_Status"].apply(lambda x: 2 if x in ["Married", "Together"] else 1)
```

Has we now have a variable expressing the household size, we could add more information by specifying the presence of a kid or a teen in the household.

```python
df_relevant["Kidhome"] = df_relevant["Kidhome"].apply(lambda x: int(x > 0))
df_relevant["Teenhome"] = df_relevant["Teenhome"].apply(lambda x: int(x > 0))
```

In the `Education` column, we find 201 instances of "2n Cycle". As this typically refers to education levels between a Bachelor's and a PhD, we label it as Master.

```python
df_relevant["Education"].value_counts()
```

```python
df_relevant["Education"] = df_relevant["Education"].replace({"2n Cycle": "Graduation", "Master": "Graduation"})
```

#### Categorical Variables

Customers with different educational backgrounds (Basic, Graduation, PhD) have distinct spending patterns on wine. Higher education levels (PhD, Master) likely correlate with higher wine spending, while lower education levels (Basic) may correlate with lower spending.

```python
def anova_test(df, var, target_var='MntWines'):
    # Prepare data for ANOVA test
    categories = df[var].unique()
    category_data = [df[df[var] == category][target_var] for category in categories]
    # ANOVA test
    anova_result = f_oneway(*category_data)
    print(f'ANOVA result for {var} and {target_var}: statistic={anova_result.statistic}, pvalue={anova_result.pvalue}')
```

```python
_,axes = plt.subplots(ncols=2, nrows=1, figsize=(10,3))
sns.boxplot(data=df_relevant, x="Marital_Status", y="MntWines", ax=axes[0])
sns.boxplot(data=df_relevant, x="Education", y="MntWines", ax=axes[1])
plt.tight_layout()
plt.show()
```

```python
anova_test(df=df_relevant, var="Marital_Status", target_var="MntWines")
anova_test(df=df_relevant, var="Education", target_var="MntWines")
anova_test(df=df_relevant, var="Income_Level", target_var="MntWines")
```

```python
def ttest_binary_var(df, var, target_var='MntWines'):
    # Prepare data for t-test
    group_0 = df[df[var] == 0][target_var]
    group_1 = df[df[var] == 1][target_var]
    # t-test
    ttest_result = ttest_ind(group_0, group_1)
    print(f't-test result for {var} and {target_var}: statistic={ttest_result.statistic}, pvalue={ttest_result.pvalue}')
```

## 5. Customer Segmentation

```python
target_vars = ["MntWines"]
numerical_vars = ["Age", "Income", "Household_Size", "TotalNumPurchases", "Customer_Tenure (days)"]
categorical_vars = ["Education"]
binary_vars = ["Kidhome"]
```

```python
df = df_relevant[target_vars + numerical_vars + categorical_vars + binary_vars].copy()
```

```python
df.info()
```

#### Scale and normalize numerical variables

```python
X = df.copy()
```

```python
def apply_boxcox(df, column):
    data_transformed, _ = boxcox(df[column]+1)
    return data_transformed
```

```python
for var in numerical_vars:
    X[var] = apply_boxcox(X, var)
```

```python
scaler = MinMaxScaler()
X[numerical_vars] = scaler.fit_transform(X[numerical_vars])
```

#### Encode categorical variables

```python
X = pd.get_dummies(X, columns=categorical_vars)
```

#### 2D Projections

```python
encoded_var_names = X.columns.tolist()
encoded_var_names.remove("MntWines")
```

```python
tsne = TSNE(n_components=2, random_state=42, init='random', learning_rate='auto')
X_tnse = tsne.fit_transform(X[encoded_var_names])
```

#### KMeans

```python
def KMeans_classification(df, min_clusters=2, max_clusters=11, figsize=(5.5, 3)):
    sse = []
    K = range(min_clusters, max_clusters+1)
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(df)
        sse.append(kmeans.inertia_)
    # Plot the elbow curve
    plt.figure(figsize=figsize)
    plt.plot(K, sse, 'o-')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.title('Elbow Method for Optimal k')
    #plt.savefig("figs/kmean_inertia.png", dpi=300, facecolor='w', bbox_inches = 'tight')
    plt.show()
```

```python
%%time
KMeans_classification(df=X, min_clusters=1, max_clusters=15, figsize=(6.5, 3.5))
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_clusters = kmeans.fit_predict(X)
np.unique(kmeans_clusters)
```

```python
print("Silhouette Score for KMeans:", silhouette_score(X, kmeans_clusters))
```

#### DBSCAN

```python
def tune_parameters_silhouette(X, epsilon_range, min_samples_values, min_clusters=2, max_clusters=10):
    best_score = -1
    best_params = {}
    best_n_clusters = 1

    # Iterate through parameter combinations
    for eps in epsilon_range:
        for min_samples in min_samples_values:
            # Instantiate DBSCAN with current parameters
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)

            # Fit DBSCAN and compute labels
            labels = dbscan.fit_predict(X) + 1
            n_clusters = np.unique(labels).shape[0]
            
            if n_clusters >= min_clusters and n_clusters <= max_clusters:
                # Compute silhouette score
                silhouette = silhouette_score(X, labels)

                # Update best score and parameters if silhouette is higher
                if silhouette > best_score:
                    best_score = silhouette
                    best_params['eps'] = eps
                    best_params['min_samples'] = min_samples
                    best_n_clusters = n_clusters

    return best_params, best_score, best_n_clusters
```

```python
%%time
best_params, best_score, n_clusters = tune_parameters_silhouette(
    X, epsilon_range=np.linspace(0.1, 2.5, num=100), 
    min_samples_values=[3, 5, 10, 20, 30, 50], 
    min_clusters=2,
    max_clusters=10
)
```

```python
print(f"Best epsilon: {best_params['eps']}")
print(f"Best min_samples: {best_params['min_samples']}")
print(f"Best silhouette: {best_score}")
print(f"Number of clusters: {n_clusters}")
```

```python
dbscan = DBSCAN(eps=best_params["eps"], min_samples=best_params["min_samples"], algorithm='auto')
dbscan_clusters = dbscan.fit_predict(X) + 1
np.unique(dbscan_clusters)
```

```python
print("Silhouette Score for DBSCAN:", silhouette_score(X, dbscan_clusters))
```

### 5.3. Clusters Analysis
* **Cluster 0** comprises individuals with lower incomes. They make fewer purchases and spend less on wine, likely due to budget constraints. Many have children at home, which might influence their purchasing decisions towards more budget-friendly options. This segment could include occasional wine buyers who may prioritize affordability.
* **Cluster 1** includes older individuals, high-income individuals who are enthusiastic about wine (spend the most on wine). They make frequent purchases and spend significantly on wine, indicating a preference for quality and variety. This segment is well-educated and typically does not have children at home, allowing for discretionary spending on premium products.
* **Cluster 2** consists of older individuals with moderate to high incomes. They make frequent purchases and have a significant but lower spending on wine compared to **Cluster 1**. This segment is also well-educated and typically child-free, allowing for regular but potentially more budget-conscious wine purchases.
