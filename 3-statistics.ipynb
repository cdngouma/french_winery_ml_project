```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

In this dataset we take a look at several datasets containing statistics about wine market in the United States.

## Gross margin of premium wineries in the United States 2002-2022

This statistic shows the gross margin of premium wineries in the United States from 2002 to 2022. According to the report, premium wineries in the United States had a gross margin of 57.5 percent as of September, 2022.

Observations:
* The gross margin of premium wineries in the US has increased over time by 11.7%.

```python
gross_margin_premium = pd.read_excel(
    "dataset/statistic_id209163_gross-margin-of-premium-wineries-in-the-united-states-2002-2022.xlsx", 
    sheet_name='Data', skiprows=4,
    names=["0:Unnamed", "Year", "Gross Margins (%)", "1:Unnamed"],
    usecols=["Year", "Gross Margins (%)"]
)
gross_margin_premium.shape
```

```python
gross_margin_premium["Year"] = gross_margin_premium["Year"].replace({"2022*": "2022"})
gross_margin_premium["Year"] = gross_margin_premium["Year"].astype(int)
```

```python
gross_margin_premium.info()
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(data=gross_margin_premium, x="Year", y="Gross Margins (%)")
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=5))
```

## Market share of the U.S. alcohol industry by beverage 2000-2022

Sales market share of the United States alcohol industry from 2000 to 2022, by beverage

Observations:
* Over the past decade, the market share of Wine has remained constant while Spirits are steadily taking more and more of the market at the expense of Beer.

```python
beverage_share = pd.read_excel(
    "dataset/statistic_id233699_market-share-of-the-us-alcohol-industry-by-beverage-2000-2022.xlsx", 
    sheet_name='Data', skiprows=4,
    names=["0:Unnamed", "Year", "Spirits", "Wine", "Beer", "1:Unnamed"],
    usecols=["Year", "Spirits", "Wine", "Beer"]
)
beverage_share.shape
```

```python
beverage_share.info()
```

```python
beverage_share = beverage_share[beverage_share["Year"] >= 2010]
```

```python
# Plotting each column as a stacked bar
plt.figure(figsize=(8,4))
plt.bar(beverage_share["Year"], beverage_share['Wine'], label='Wine')
plt.bar(beverage_share["Year"], beverage_share['Spirits'], bottom=beverage_share['Wine'], label='Spirits')
plt.bar(beverage_share["Year"], beverage_share['Beer'], bottom=beverage_share['Wine'] + beverage_share['Spirits'], label='Beer')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=5))
plt.legend()
```

## Sales volume of wine in the U.S. 1999-2021

Sales volume of wine in the United States from 1999 to 2021 ( in million 9-liter cases)

Observations:
* Wine sales have consistently increased over the year and the trend is still going.

```python
sales_volume = pd.read_excel(
    "dataset/statistic_id715474_sales-volume-of-wine-in-the-us-1999-2021.xlsx", 
    sheet_name='Data', skiprows=4,
    names=["0:Unnamed", "Year", "Volume (in million 9-liter cases)"],
    usecols=["Year", "Volume (in million 9-liter cases)"]
)
sales_volume.shape
```

```python
sales_volume.info()
```

```python
def convert_cases_to_gallons(million_cases):
    gallons_per_case = 2.37754858  # Conversion factor from 9-liter case to gallons
    million_gallons = million_cases * gallons_per_case
    return million_gallons
```

```python
sales_volume["Volume (in million gallons)"] = sales_volume["Volume (in million 9-liter cases)"].apply(convert_cases_to_gallons)
```

```python
plt.figure(figsize=(7,4))
sns.lineplot(data=sales_volume, x="Year", y="Volume (in million gallons)")
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=5))
```

## Combined Dataset

We combine the different datasets into a single one on the `Year` column. The data thus goes from 2010 to 2021. Given the context of our study we could consider the following columns are target:
* `Premium Wineries Gross Margins (%)`
* `Californian Wine Market Value ($Bn)`
* `US Wine Sales Volume (Mgal)`

Observations:
* Some variables are strongly correlated with each other.
* `Californian Wine Market Value`, `Premium Wineries Gross Margins`, `US Wine Consumption`, and `US Wine Sales Volume` are strongly correlated with each other.
* `Californian Wine Production` is strongly correlated with `US Wine Production`.
* `Premium Wineries Gross Margins` is also correlated tho weaker with `US Wine Production` and `Californian Wine Production`.
* `Wine Market Share` also has some correlating with `US Wine Production` and `Californian Wine Production`.

```python
merged_df = pd.merge(
    gross_margin_premium.rename(columns={"Gross Margins (%)": "Premium Wineries Gross Margins (%)"}), 
    beverage_share[["Year", "Wine"]].rename(columns={"Wine": "US Wine Market Share (%)"}), 
    on='Year', how="outer"
)
```

```python
merged_df = pd.merge(
    merged_df, 
    wine_consumption.rename(columns={"Consumption (in million gallons)": "US Wine Consumption (Mgal)"}), 
    on='Year', how='outer'
)
```

```python
merged_df = pd.merge(
    merged_df, 
    wine_production.rename(columns={
        "US (in million gallons)": "US Wine Production (Mgal)", 
        "California (in million gallons)": "California Wine Production (Mgal)"
    }), 
    on='Year', how='outer'
)
```

```python
merged_df = pd.merge(
    merged_df, 
    sales_volume.rename(columns={"Volume (in million gallons)": "US Wine Sales Volume (Mgal)"})[["Year", "US Wine Sales Volume (Mgal)"]], 
    on='Year', how='outer'
)
```

```python
merged_df = pd.merge(
    merged_df, 
    californian_market_value.rename(columns={"Value (in billion U.S. dollars)": "Californian Wine Market Value ($Bn)"}), 
    on='Year', how='outer'
)
```

```python
merged_df = merged_df.sort_values(by="Year")
```

```python
def calculate_correlation(dataframe):
    # Calculate correlation matrix for the given DataFrame
    correlation_matrix = dataframe.corr()
    return correlation_matrix
```

```python
def visualize_correlation(correlation_matrix):
    # Visualize the correlation matrix using a heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt='.2f', linewidths=.5)
    plt.title('Correlation Between Variables')
    #plt.xlabel('Time Series')
    #plt.ylabel('Time Series')
    plt.savefig("correlation_table.png")
```

```python
correlation_matrix = calculate_correlation(merged_df.drop(columns=["Year"]))
```

```python
visualize_correlation(correlation_matrix)
```
